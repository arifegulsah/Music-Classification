## 1. Importing Libs for Preprocessing

Ses verilerimizi işleyebilmek için *librosa* paketimizden faydalanacağız.<br>
(Çıkan warningi görmezden gelebilirsiniz.)

```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import csv
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

## 2. Uploading & Preprocessing Data

Verilerimizi modelde eğitmeden önce ön işlemesini yapmak durumundayız.<br>
Bunun için ses verilerinde uygulanabilecek belli başlı yöntemler bulunmaktadır. Bu yöntemlerden bazıları: <br>
1. Kromagramın hesaplanması,
2. Spektrogramdan her kare için ortalama karekökün hesaplanması,
3. Spektral sentroidin hesaplanması,
4. Spektral bant genişliğinin hesaplanması,
5. Devre dışı bırakma frekansının hesaplanması,
6. Sıfır geçiş oranının hesaplanması.

Daha detaylı bilgiyi librosa paketinin [dokümantasyonu](https://librosa.org/doc/latest/index.html) okuyarak edinebilirsiniz.
<br><br>
Ses verilerimizin özniteliklerini çıkardıktan sonra elde edecek olduğumuz değerleri incelemek ve üzerlerinde işlem yapabilmek için manuel şekilde bir csv uzantılı dosya oluşturuyoruz.
<br>
Oluşturduğumuz csv dosyasında sütun isimleri olarak; *filename* adı altında ses dosyalarımızın isimleri ve yukarıda bahsi geçen yöntemler için özel adlandırmaların adı altında da sonuç olarak elde edeceğimiz değerlerimiz yer alacaktır.
<br>
Bu kısımlar bizim **featurelarımızın** ta kendisidir.
<br>
*label* isimli sütunda ise **classlarımız** bulunacaktır.
<br>
*csvfilepath* değişkenimize de ses verilerimizin olduğu dosyanın path'ini veriyoruz. Bu sayede verilerimizi yükleyeceğiz.

```python
header = 'filename chroma_stft rmse spectral_centroid spectral_bandwidth rolloff zero_crossing_rate'
for i in range(1, 21):
    header += f' mfcc{i}'
header += ' label'
header = header.split()

csvfilepath = 'csvfile/audioDatas.csv'
```

Şimdi sıra oluşturduğumuz csv dosyasını doldurmakta.
<br>
Bunun için dosyamızı açıyoruz ve aşağıda verdiğimiz döngü aracılığı ile bütün ses dosyalarımıza öznitelik çıkarımı yapıyoruz. 
<br>
Değerlerimiz teker teker dosyamıza işleniyor. Bu işlem biraz uzun sürebilir.

```python
file = open(csvfilepath, 'w', newline='')
with file:
    writer = csv.writer(file)
    writer.writerow(header)
genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()
for g in genres:
    for filename in os.listdir(f'genres/{g}'):
        songname = f'genres/{g}/{filename}'
        y, sr = librosa.load(songname, mono=True, duration=30)
        chroma_stft = librosa.feature.chroma_stft(y=y, sr=sr)
        rmse = librosa.feature.rms(y=y)
        spec_cent = librosa.feature.spectral_centroid(y=y, sr=sr)
        spec_bw = librosa.feature.spectral_bandwidth(y=y, sr=sr)
        rolloff = librosa.feature.spectral_rolloff(y=y, sr=sr)
        zcr = librosa.feature.zero_crossing_rate(y)
        mfcc = librosa.feature.mfcc(y=y, sr=sr)
        to_append = f'{filename} {np.mean(chroma_stft)} {np.mean(rmse)} {np.mean(spec_cent)} {np.mean(spec_bw)} {np.mean(rolloff)} {np.mean(zcr)}'    
        for e in mfcc:
            to_append += f' {np.mean(e)}'
        to_append += f' {g}'
        file = open(csvfilepath, 'a', newline='')
        with file:
            writer = csv.writer(file)
            writer.writerow(to_append.split())
```

İşlem bittikten sonra dosyamızı yani yeni veri setimizi inceleyelim.

```python
data = pd.read_csv(csvfilepath)
data.head()
```

Görüldüğü üzere ses dosyalarımız için öznitelikler çıkarılmış ve sayısal değerler olarak kaydedilmiş. Label sütunun altında ise o ses dosyasının ait olduğu sınıf ismi bulunmakta.
<br>
Modeli eğitirken bu sayısal değerlerden yararlanacağız, lakin dosya isimlerinin tutulduğu *filename* sütunu modeli eğitirken hiçbir işe yaramayacaktır. O sütunu droplayarak devre dışı bırakıyoruz.

```python
data = data.drop(['filename'],axis=1)
data.head()
```

Veri önişleme adımlarından birisi olan "label encoder" olayı kategorik olan her verinin sayısallaştırılmasına denir. Bizim veri setimizdeki kategorik tipindeki tek sütun *label* isimli sütunumuz.
<br>
iloc fonksiyonun ikinci parametresine -1 vererek sondaki label isimli sütunu ele alıyoruz ve fit transformdan geçiriyoruz.
<br>
Sonucu yazdırdığımız zaman bütün class isimlerimizin ***(blues, classical, country, disco, hiphop, jazz, metal, pop, reggae ve rock)***, 0 ile 9 arasında numaralara atandığını görüyoruz.

```python
genre_list = data.iloc[:, -1]
encoder = LabelEncoder()
y = encoder.fit_transform(genre_list)
print(y)
```

Veri önişleme adımlarından sonuncusu olan öznitelik ölçeklendirme (feature scaling) işlemimizi de gerçeleştiriyoruz.
<br>
Bildiğimiz üzere bunu yapmamızın sebebi modelin büyük aralıklı değişkenlerin küçük aralıklı değişkenlerden daha önemli olduğunu düşünmesini engellemek.
<br>
<br>
iloc fonksiyonun :-1 diyerek class'larımızın bulunduğu sütün hariç bütün sütunlara feature scaling işlemi gerçekleştiriyoruz. Sonra bu değerleri X ismini verdiğimiz değişkende turuyoruz.
<br>
Classlarımızı bir önceki adında y isimli değişkende zaten tutmuştuk.
<br>
X ve y değerlerimizi train ve test şeklinde ayırmak için train_test_split fonksiyonunu kullanıyoruz.
<br>
Bu sayede ön işleme aşamamız bitmiş oldu.

```python
scaler = StandardScaler()
X = scaler.fit_transform(np.array(data.iloc[:, :-1], dtype = float))

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 3. Importing Libs for Model Building

Modelimizi oluşturuken *keras* kütüphanesinden faydalanacağız.

```python
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras import models
from tensorflow.keras import layers
```

## 4. Building Model

Sequential ile modelimizi oluşturuyoruz. <br>
add layer ile katmanlarını ekliyoruz ve compile ile de parametrelerini veriyoruz.

```python
model = models.Sequential()
model.add(layers.Dense(256, activation='relu', input_shape=(X_train.shape[1],)))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```


Artık modelimizi eğitebiliriz. Bunun için fit fonksiyonundan faydalanıyoruz.

```python
history = model.fit(X_train,
                    y_train,
                    epochs=200,
                    batch_size=256)
```

## 5. Evaluating Model

```python
test_loss, test_acc = model.evaluate(X_test,y_test)
print('test_acc: ',test_acc)
```

Tahminlerimizi *predictions* ismini verdiğimiz değişkende tutuyoruz. 
<br>
Bu değişkeni incelediğimiz zaman her bir ses dosyasının, yine her bir müzik türlerine olan yakınlığını görmekteyiz.
<br>
Burada bizim için önemli olan şey, modelin hesapladığı en yakın türü bulmaktır. Bunu argmax kullanarak yapabiliriz. 
<br>
Örneğin burada 3. ses dosyası için tahmin edilen sınıfı incelediğimiz zaman 8 numaralı sınıf olarak hesaplandığını görebiliyoruz.

```python
predictions = model.predict(X_test)
np.argmax(predictions[2])
```

Bütün satırlarımız için bu işlemi uygulayalım.
<br>
Bunun için aşağıdaki for döngüsünü kullanabiliriz.
<br>
y_pred ismini verdiğimiz yeni değişkende artık modelin tahmin ettiği sınıflar bulunmakta.
<br>
Bunu orjinal sonuçlar ile karşılaştırarak modelimizin ne kadar iyi bir öğrenme gerçekleştirdiğini artık anlayabiliriz.

```python
maxpredicts = []
for element in predictions:
    temp = np.argmax(element)
    maxpredicts.append(temp)
    
y_pred = np.array(maxpredicts)    
```

Sonuçlarımızı kolayca değerlendirmek için classificaiton_report'dan yararlancağız. 
<br>
classification_report bize precision, recall ve f1 score gibi metrikleri döndüren bir fonskiyondur.
<br>
Bunun için öncelikle class isimlerimi target_names isimli bir değişkende tutuyorum. Ve fonksiyonumu y_test (orjinal sonuçlar) ve y_pred (modelin tahmin ettiği sonuçlar) olarak çağırıyorum.

```python
target_names = ['0blues', '1classical', '2country', '3disco', '4hiphop', '5jazz', '6metal', '7pop', '8reggae', '9rock']
```

```python
from sklearn.metrics import classification_report
classification_report(y_test, y_pred, target_names=target_names)
```

Fonksiyonum başarılı bir şekilde çalıştı ve bana sonuçları döndürdü. Lakin görünüşü pek hoş değil ve kafa karıştırıcı. Bu yüzden bu sonuçlarımı daha düzenli gözükmelerini sağlamak için yeni bir csv filea taşımak istiyorum.
<br>
Bunun için tek yapmam gereken:

```python
clsf_report = pd.DataFrame(classification_report(y_true = y_test, y_pred = y_pred,target_names=target_names, output_dict=True)).transpose()
clsf_report.to_csv('csvfile/classificationReport.csv', index= True)
```
